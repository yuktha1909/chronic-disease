# Diabetes Dataset EDA

This notebook performs basic exploratory analysis to support the final-year project report.


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from config import DATASET_PATH, ROOT_DATASET_FALLBACK_PATH, DATASET_URL

try:
    df = pd.read_csv(DATASET_PATH)
except FileNotFoundError:
    try:
        df = pd.read_csv(ROOT_DATASET_FALLBACK_PATH)
    except FileNotFoundError:
        df = pd.read_csv(DATASET_URL)

df.head()
```

```python
df.info()
```

```python
corr = df.corr(numeric_only=True)
plt.figure(figsize=(10, 7))
sns.heatmap(corr, cmap='RdBu_r', center=0)
plt.title('Feature Correlation Matrix')
plt.show()
```
